2. **Structuring**

```python
df.columns = df.columns.str.strip()
df.columns
```

(ii) What is the seasonal performance of each crop, in terms of both total Production and

```python
seasonal_performance=df.groupby(['Crop','Season']).agg(total_Production=('Production','sum'),Average_Yield=('Yield','mean'))
seasonal_performance
```

```python
yearwise_total_production=df.pivot_table(index='Crop',columns='Crop_Year',values='Production',aggfunc='sum')
yearwise_total_production
```

(iv) Which crop, in which year, produced the highest average yield

3. **Cleaning**

(i) Change inconsistent columns values like 'State' and 'District' from uppercase to
lowercase and remove trailing or leading spaces


```python
df['State'] = df['State'].str.lower().str.strip()
df['District'] = df['District'].str.lower().str.strip()
df['District'].unique()
```

(ii) Instead of filling missing Production with 0 or a global mean, impute the NaN values
by calculating Area * (Average_Yield). The Average_Yield should be the average for
that specific District and Crop combination.


```python
df['Average_Yield'] = df.groupby(['District', 'Crop'])['Yield'].transform('mean')
df['Production'] = df['Production'].fillna(df['Area'] * df['Average_Yield'])
df.head()
```

(iii) Find the wheat production of Odisha state


```python
odisha_wheat = df[(df['State'] == 'odisha') & (df['Crop'] == 'wheat')]
print(odisha_wheat['Production'].sum())
```

4. **Enrichment**

(i) Merge with External Data


```python
df1=pd.read_csv('Crop_Production1.csv')
df2=pd.read_csv('city_day.csv')
df1.rename(columns={'District ':'City'},inplace=True)
all_names=set(df1['City'])|set(df2['City'])
sorted(all_names)

dict={'AHMADABAD':'ahmedabad','BANGALORE RURAL':'Bengaluru','CHENNAI':'Chennai','DELHI_TOTAL':'Delhi'}
df1['City']=df1['City'].replace(dict)
df1['City']=df1['City'].str.lower().str.strip()

df2['City']=df2['City'].replace(dict)
df2['City']=df2['City'].str.lower().str.strip()

df3=pd.merge(df1,df2,how='left',on='City')
df3.head()
```

(ii) Create new column {Pollutant ratios} derived from existing columns {SO2,NOx}

```python
df3['Pollutant_Ratio'] = df3['SO2'] / df3['NOx']
df3['Pollutant_Ratio'].head()
```

5. **Validating**

(i) The Yield column should equal Production / Area. Create a new column called
Calculated_Yield using this formula. Then, find the sum of the difference between
Calculated_Yield and Yield to see if the original data was accurate





```python
df3['Calculated_Yield'] = df3['Production'] / df3['Area ']
sum_diff = (df3['Calculated_Yield'] - df3['Yield']).sum()
print("Total difference in yield:", sum_diff)
```
